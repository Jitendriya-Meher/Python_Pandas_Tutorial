```python
df = pd.read_csv('weather_data.csv',parse_dates=['day'])
print(type(df.day[0]))
df
```

fillna

Fill all NaN with one specific value

```python
df.fillna(0)
```

Fill na using column names and dict

```python
df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':"No Event"
})
```

Use method to determine how to fill na values

forward fill - fill from upward

```python
new_df = df.fillna(method="ffill")
new_df
```

backword fill - fill from bottom

```python
new_df = df.fillna(method="bfill")
new_df
```

```python
new_df = df.fillna(method="ffill",axis="columns")
new_df
```

```python
new_df = df.fillna(method="bfill",axis="columns")
new_df
```

```python
new_df = df.fillna(method="bfill",limit=1)
new_df
```

interpolate

```python
new_df = df.interpolate()
new_df
```

dropna

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
# all cloumn should be zero 
new_df
```

df.dropna(thresh=2): Rows with at least 2 non-NaN (non-missing) values remain in the DataFrame.

```python
new_df = df.dropna(thresh=2)
new_df
```

```python
new_df = df.dropna(thresh=3)
new_df
```
